```python
# Import necessary libraries
import os
import sys
import joblib  # Assuming models are saved using joblib
import pandas as pd  # For DataFrame and Excel handling
import json  # For saving results to JSON

# Add the scripts directory to the Python path
scripts_path = os.path.abspath(os.path.join(os.getcwd(), "../scripts"))
if scripts_path not in sys.path:
    sys.path.append(scripts_path)

from Evaluation import evaluate_model  # Import the function from Evaluation.py

# Paths
models_dir = os.path.abspath(os.path.join(os.getcwd(), "../models"))  # Ensure this path is correct

# Initialize results list
results = []

# Iterate over model files in the directory
for model_file in os.listdir(models_dir):
    if model_file.endswith(".pkl"):  # Check for .pkl files
        model_path = os.path.join(models_dir, model_file)
        model_name = os.path.splitext(model_file)[0]

        # Load the model
        with open(model_path, 'rb') as file:
            try:
                model = joblib.load(file)
                print(f"Evaluating model: {model_name}")

                # Evaluate the model
                metrics = evaluate_model(model_name, model)
                if metrics:
                    # Append the model name and metrics to the results list
                    results.append({"Model Name": model_name, **metrics})

            except Exception as e:
                print(f"Error loading or evaluating model {model_name}: {e}")

# Display results
print("\nEvaluation Metrics:")
for result in results:
    print(result)

# Convert results to a DataFrame for Excel saving
df = pd.DataFrame(results)

# Save the DataFrame to an Excel file
excel_path = os.path.join(os.getcwd(), "../results/metrics/model_evaluation_metrics.csv")
df.to_csv(excel_path, index=False)

print(f"\nEvaluation Metrics saved to {excel_path}")

# Save the results to a JSON file
json_path = os.path.join(os.getcwd(), "../results/metrics/model_evaluation_metrics.json")
with open(json_path, 'w') as json_file:
    json.dump(results, json_file, indent=4)

print(f"Evaluation Metrics saved to {json_path}")
```
